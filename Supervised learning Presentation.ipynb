**Importing libraries for supervised learning**

```python
import sklearn as sk
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Setting Manual Randomness
np.random.seed(42)

# Creating Values
X = np.linspace(start = 0, stop = 10, num = 50)
y = 2 * X + np.random.randint(low = -4 , high = 4, size = len(X))
```

```python
# Visualizing the Values
plt.scatter(X, y)
plt.show()
```

```python
# Fitting the model

model = LinearRegression(fit_intercept = True)
X = np.expand_dims(X, axis = 1)
model.fit(X, y)
print(f"Linear Slope: {model.coef_}\nLinear Intercept: {model.intercept_}")
```

```python
# Testing model on new data
X_new = np.linspace(start = -5, stop = 15, num = 20)
y_pred = model.predict(np.expand_dims(X_new, axis = -1))

# Visualizing predicted data
plt.scatter(X, y)
plt.scatter(X_new, y_pred)
plt.plot(X_new, y_pred, c = "red")
```

**Decision Trees**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.tree import plot_tree
model_2 = DecisionTreeClassifier()
iris = load_iris()
X, y = iris.data, iris.target
```

```python
model_2.fit(X, y)
plt.figure(figsize = (16,10))
plot_tree(model_2, filled = True, proportion = True)
```

```python
print(f"Predicted desicion for X[20]: {model_2.predict([X[20]])}")
print(f"True value decision for X[20]: {y[20]}")
```
